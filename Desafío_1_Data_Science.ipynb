### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda.columns = [col.replace(' ', '_') for col in tienda.columns]
print (tienda.columns)
tienda2.columns = [col.replace(' ', '_') for col in tienda2.columns]
print (tienda2.columns)
tienda3.columns = [col.replace(' ', '_') for col in tienda3.columns]
print (tienda3.columns)
tienda4.columns = [col.replace(' ', '_') for col in tienda4.columns]
print (tienda4.columns)
```

#1. Análisis de facturación



```python
#Ventas por tienda
ventas_tienda1 = tienda['Precio'].sum()
ventas_tienda2 = tienda2['Precio'].sum()
ventas_tienda3 = tienda3['Precio'].sum()
ventas_tienda4 = tienda4['Precio'].sum()

print (f'Las ventas totales de la tienda 1 son: {ventas_tienda1}')
print (f'Las ventas totales de la tienda 2 son: {ventas_tienda2}')
print (f'Las ventas totales de la tienda 3 son: {ventas_tienda3}')
print (f'Las ventas totales de la tienda 4 son: {ventas_tienda4}')
```

# 2. Ventas por categoría

```python
#Ventas por categoría tienda 1
ventas_categoria1 = tienda.groupby('Categoría_del_Producto')['Precio'].count()
ventas_categoria1 = ventas_categoria1.sort_values(ascending=False)
print (ventas_categoria1.head(4))
print (ventas_categoria1.tail(4))
#Ventas por categoría tienda 2
ventas_categoria2 = tienda2.groupby('Categoría_del_Producto')['Precio'].count()
ventas_categoria2 = ventas_categoria2.sort_values(ascending=False)
print (ventas_categoria2.head(4))
print (ventas_categoria2.tail(4))
#Ventas por categoría tienda 3
ventas_categoria3 = tienda3.groupby('Categoría_del_Producto')['Precio'].count()
ventas_categoria3 = ventas_categoria3.sort_values(ascending=False)
print (ventas_categoria3.head(4))
print (ventas_categoria3.tail(4))
#Ventas por categoría tienda 4
ventas_categoria4 = tienda4.groupby('Categoría_del_Producto')['Precio'].count()
ventas_categoria4 = ventas_categoria4.sort_values(ascending=False)
print (ventas_categoria4.head(4))
print (ventas_categoria4.tail(4))
```

# 3. Calificación promedio de la tienda


```python
#Calificación promedio tienda 1
calificacion_tienda1 = round(tienda['Calificación'].mean(),2)
print (f'La calificación promedio de la tienda 1 es : %{calificacion_tienda1}')
#Calificación promedio tienda 2
calificacion_tienda2 = round(tienda2['Calificación'].mean(),2)
print (f'La calificación promedio de la tienda 2 es : %{calificacion_tienda2}')
#Calificación promedio tienda 3
calificacion_tienda3 = round(tienda3['Calificación'].mean(),2)
print (f'La calificación promedio de la tienda 3 es : %{calificacion_tienda3}')
#Calificación promedio tienda 4
calificacion_tienda4 = round(tienda4['Calificación'].mean(),2)
print (f'La calificación promedio de la tienda 4 es : %{calificacion_tienda4}')
```

# 4. Productos más y menos vendidos

```python
#Ventas por producto tienda 1
ventas_producto1 = tienda.groupby('Producto')['Precio'].count()
ventas_producto1 = ventas_producto1.sort_values(ascending=False)
print (ventas_producto1.head(5))
print (ventas_producto1.tail(5))
#Ventas por producto tienda 2
ventas_producto2 = tienda2.groupby('Producto')['Precio'].count()
ventas_producto2 = ventas_producto2.sort_values(ascending=False)
print (ventas_producto2.head(5))
print (ventas_producto2.tail(5))
#Ventas por producto tienda 3
ventas_producto3 = tienda3.groupby('Producto')['Precio'].count()
ventas_producto3 = ventas_producto3.sort_values(ascending=False)
print (ventas_producto3.head(5))
print (ventas_producto3.tail(5))
#Ventas por producto tienda 4
ventas_producto4 = tienda4.groupby('Producto')['Precio'].count()
ventas_producto4 = ventas_producto4.sort_values(ascending=False)
print (ventas_producto4.head(5))
print (ventas_producto4.tail(5))
```

# 5. Envío promedio por tienda

```python
#Envío promedio tienda 1
envío_tienda1 = round(tienda['Costo_de_envío'].mean(),2)
print (f'El envío promedio de la tienda 1 es : {envío_tienda1}')
#Envío promedio tienda 2
envío_tienda2 = round(tienda2['Costo_de_envío'].mean(),2)
print (f'El envío promedio de la tienda 2 es : {envío_tienda2}')
#Envío promedio tienda 3
envío_tienda3 = round(tienda3['Costo_de_envío'].mean(),2)
print (f'El envío promedio de la tienda 3 es : {envío_tienda3}')
#Envío promedio tienda 4
envío_tienda4 = round(tienda4['Costo_de_envío'].mean(),2)
print (f'El envío promedio de la tienda 4 es : {envío_tienda4}')
```

```python
#gastos por envíos por tienda

gastos_tienda1 = tienda['Costo_de_envío'].sum()
gastos_tienda2 = tienda2['Costo_de_envío'].sum()
gastos_tienda3 = tienda3['Costo_de_envío'].sum()
gastos_tienda4 = tienda4['Costo_de_envío'].sum()

print (f'Los gastos totales de la tienda 1 son: {gastos_tienda1}')
print (f'Los gastos totales de la tienda 2 son: {gastos_tienda2}')
print (f'Los gastos totales de la tienda 3 son: {gastos_tienda3}')
print (f'Los gastos totales de la tienda 4 son: {gastos_tienda4}')
```

```python
#Utilidad neta por tienda
utilidad_tienda1 = ventas_tienda1 - gastos_tienda1
utilidad_tienda2 = ventas_tienda2 - gastos_tienda2
utilidad_tienda3 = ventas_tienda3 - gastos_tienda3
utilidad_tienda4 = ventas_tienda4 - gastos_tienda4

print (f'La utilidad neta de la tienda 1 es: {utilidad_tienda1}')
print (f'La utilidad neta de la tienda 2 es: {utilidad_tienda2}')
print (f'La utilidad neta de la tienda 3 es: {utilidad_tienda3}')
print (f'La utilidad neta de la tienda 4 es: {utilidad_tienda4}')
```

```python
#gráfico utilidad
Tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
Utilidad = [utilidad_tienda1,utilidad_tienda2,utilidad_tienda3,utilidad_tienda4]
grafico_utilidad = plt.barh(Tiendas, Utilidad)
plt.title('UTILIDAD NETA POR TIENDA')
plt.xlabel('Tiendas')
plt.ylabel('Utilidad')
```
